```python

import pandas as pd
dfbank=pd.read_csv('https://raw.githubusercontent.com/jarsmp37/introds/main/Churn_Bank.csv')
dfbank
```

```python
dfbank1=dfbank.dropna()
```

```python
columnas=['RowNumber', 'CustomerId', 'Surname', 'CreditScore', 'Geography',
       'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary', 'Exited', 'Complain',
       'Satisfaction Score', 'Card Type', 'Point Earned']
```

```python

dfbank2 = dfbank1
for v in columnas:
  dfbank2 = dfbank2[dfbank2[v]!='invalid_value']
```

```python
dfbank2.isnull().sum()
```

```python


for c in columnas:
    print(f"{c}: {dfbank2[dfbank2[c]=='invalid_value'].shape[0]}")
```

```python
dfbank2
```

```python
dfbank3 = dfbank2.drop_duplicates(subset=['CustomerId'])
dfbank3
```

```python
dfbank3=dfbank3.rename(columns={'Surname': 'Apellido','Geography': 'Pais','Gender': 'Género','Age': 'Edad','Tenure': 'Estadía','NumOfProducts': '#Productos','HasCrCard': 'Tarjeta_Credito','IsActiveMember': 'Activo','EstimatedSalary': 'Salario', 'Exited': 'Salida', 'Complain': 'Quejas', 'Satisfaction Score': 'Nivel_Satisfacción', 'Card Type': 'Tipo_Tarjeta', 'Point Earned': 'Puntos'})
```

```python
dfbank3
```

```python
dfbank3['Pais'].unique()
```

```python

traduccion={
   'Spain':'España',
   'France':'Francia',
   'Germany':'Alemania'
}
```

```python

dfbank3["Pais"]=dfbank3["Pais"].replace(traduccion)
```

```python
dfbank3['Género'].unique()
```

```python


traduccion2={
   'Female':'Mujer',
   'Male':'Hombre'
}
```

```python


dfbank3["Género"]=dfbank3["Género"].replace(traduccion2)
dfbank3
```

```python
dfbank3['CustomerId']=dfbank3['CustomerId'].astype(float).astype(int)
dfbank3['Edad']=dfbank3['Edad'].astype(float).astype(int)
dfbank3['Estadía']=dfbank3['Estadía'].astype(float).astype(int)
dfbank3['Nivel_Satisfacción']=dfbank3['Nivel_Satisfacción'].astype(float).astype(int)
dfbank3['Puntos']=dfbank3['Puntos'].astype(float).astype(int)
dfbank3['Salida']=dfbank3['Salida'].astype(float).astype(int)
dfbank3['Quejas']=dfbank3['Quejas'].astype(float).astype(int)
dfbank3['#Productos']=dfbank3['#Productos'].astype(float).astype(int)
dfbank3['Tarjeta_Credito']=dfbank3['Tarjeta_Credito'].astype(float).astype(int)
dfbank3['Activo']=dfbank3['Activo'].astype(float).astype(int)
```

```python
dfbank3['Balance']=dfbank3['Balance'].astype(float)
dfbank3['Salario']=dfbank3['Salario'].astype(float)
```

```python
dfbank3.info()
```

```python
dfbank4=dfbank3.drop(columns=['RowNumber'])
```

```python
dfbank4=dfbank4.reset_index(drop=True)
```
