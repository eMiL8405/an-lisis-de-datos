```python
import pandas as pd
dfbank=pd.read_csv('https://raw.githubusercontent.com/Reyvid-byte/Bank/refs/heads/main/Base_Chunk_Limpia.csv')
dfbank
```

```python
#Crear el histograma
plt.figure(figsize=(8,6))
plt.hist(dfbank['Edad'], bins=74, color='purple', edgecolor='black');

#Personalizar
plt.title('Histograma de las edades', fontsize=16)
plt.xlabel('Edades', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

plt.show()
```
